# Simulating the temperature in the lithosphere

### Defines the function that solves the diffusion equation with a heat-source. 
#### We have to update the forward euler scheme slightly to take care of the heat-sources. The steps are explained in the report. We now have an scaled equation, where we can tweak $C_1$ and $C_2$ according to the boundary condtions and the heat source. 

```python
import numpy as np
import matplotlib.pyplot as plt
from numba import jit
import time

def forward_euler(x_min, x_max, dx, dt, t, g, a, b, C_1, C_2):
    """Solves the heat equation using forward euler.

    Function for solving the diffusion equation in one dimension with
    given boundary conditions and added radioactive elements.

    Parameters
    ----------
    x_min: float
        Lower limit of spacial dimension.
    x_max: float
        Upper limit of spacial dimension.
    dx: float
        Spacing between spacial points.
    dt: float
        Spacing between time steps.
    t: float
        max time step.
    g: function(x)
        Initial condition for u(x,0).
    a: function(t)
        Boundary condition u(0,t)
    b: function(t)
        Boundary condition u(x_max,t)
    C_1: float
        Constant in the scaled equation. 
    C_2: function(y)
        Piecewise constant in the scaled equation. 

    Returns
    -------
    solutions: numpy matrix
        Contains the solution for each time step up to t.
        First dimension: Time steps
        Second dimension: Spacial steps.
    """

    num_t_values = int(float(t)/dt + 1)
    num_x_values = int(float(x_max - x_min)/dx + 1)

    #Matrix the will contain the solution for each time-step.
    u = np.zeros((num_t_values, num_x_values))

    #Vectors for x-values and t_values in the grid.
    x_values = np.linspace(x_min, x_max, num_x_values)
    t_values = np.linspace(0, t, num_t_values)

    alpha = dt/(dx**2)

    #Loops over the u-matrix and sets the boundary conditions.
    for t_index in range(0, num_t_values):
        u[t_index, 0] = a(t_values[t_index])
        u[t_index, -1] = b(t_values[t_index])

    #Loops over the u-matrix and sets the initial conditions.
    for x_index in range(1, num_x_values-1):
        u[0, x_index] = g(x_values[x_index])

    start_time = time.time()
    u = update_forward_euler(t_values, x_values, u, alpha, dt, C_1, C_2)
    end_time = time.time()

    print("Time used forward Euler: ", end_time - start_time, " seconds")

    return u

# @jit(nopython=True)
def update_forward_euler(t_values, x_values, u, alpha, dt, C_1, C_2 ):
    """Function for finding the u-matrix in forward euler with added heat.

    Parameters
    ----------
    num_t_values: int
        Number of t-values in the grid.
    num_x_values: int
        Number of x-values in the grid.
    u: numpy 2d-array
        Empty matrix that will contain the solution for each time-step.
    alpha: float

    Returns
    -------
    u: numpy 2d-array
        Matrix that contains the solution for each time-step.
    """
    num_t_values = len(t_values)
    num_x_values = len(x_values)
    for t_index in range(1, num_t_values):
        for x_index in range(1, num_x_values-1):
            u[t_index, x_index] = C_1*alpha*(u[t_index-1, x_index + 1] - 2*u[t_index-1, x_index] + u[t_index-1, x_index - 1]) + u[t_index-1, x_index] + dt*C_2(x_values[x_index], t=t_values[t_index])
    return u
```

### Sets up the parameters in the model and find the characteristic time and depth. Uses 120km for depth and 1Gy for time. Sets time = 0 to be 1Gy ago. 

```python
#PARAMETERS
rho = 3.5*10**3
c_p = 1000
k = 2.5
alpha = k/(rho*c_p)

T_1 = 1300
T_0 = 8

#Characteristic sizes.
y_c = 120*10**3 #meters
t_c = (10**9)*365*24*60*60 #seconds

C_1 = (t_c*alpha)/(y_c**2)
```

## Finds the steady state before radioactive enrichment. This is done by setting Q = 0. 

```python
g = (lambda x : 0)
a = (lambda t : 1)
b = (lambda t : 0)

def Q_2(y_bar, additional_heat =0, t=0):
    return 0

def C_2(y_bar, add_heat = 0.5*10**(-6), t = 0):
    y = y_bar
    temp = ((t_c)/((rho*c_p)*(T_1-T_0)))*Q_2(y,add_heat, t=t)
    return temp

def analytic_no_heat(y_bar):
    return 1-y_bar

y_bar_min = 0
y_bar_max = 1
dy_bar = 0.02
dt_bar = 0.00005
t = 1

num_y_bar_values = (y_bar_max - y_bar_min)/dy_bar + 1

y_bar_values = np.linspace(y_bar_min, y_bar_max, num_y_bar_values)
y_values = y_bar_values*120

T_bar = forward_euler(y_bar_min, y_bar_max, dy_bar, dt_bar, t, g, a, b, C_1, C_2)
T_bar_analytic = (analytic_no_heat(y_bar_values))

T = T_bar*(T_1 - T_0) + T_0
T_analytic = T_bar_analytic*(T_1-T_0) + T_0

y_values = y_values - 120

plt.figure(figsize = (15,8))


plt.plot(y_values, T[-1,:], alpha=0.9, linestyle="--", marker = "x", ms=3)
plt.plot(y_values, T_analytic, alpha=0.9, linestyle="--", marker = "x", ms=3)


plt.title("Steady state temperature before enrichment", fontsize=20)

plt.ylabel(r"T $[^\circ C]$", fontsize=20)
plt.xlabel("y [km]", fontsize=20)

plt.legend(["Numerical", "Analytic"], fontsize=20)

plt.savefig(r"Plots/steady_state_lithospere.png")

plt.show()
```

## Simulates the temperature from 1Gy ago until today with radioactive enrichment (No additional radioactive enrichment for now). Assumes that the initial condition is the linear function found as the steady state in the previous excercise. 

```python
g = (lambda x : 1-x)
a = (lambda t : 1)
b = (lambda t : 0)

def Q_1(y_bar, additional_heat =0, t=0):
    y = y_bar
    additional_heat = additional_heat*(0.4*np.exp(-t/4.47) + 0.4*np.exp(-t/14) + 0.2*np.exp(-t/1.25))
    if y < 0 or y > 120/120:
        print("Error out of bounds")
    elif 0 <= y < 80/120:
        return 0.05*(10**(-6)) + additional_heat
    elif 80/120 <= y < 100/120:
        return 0.35*(10**(-6))
    else:
        return 1.4*(10**(-6))


def C_2(y_bar, add_heat = 0.5*10**(-6), t = 0):
    y = y_bar
    temp = ((t_c)/((rho*c_p)*(T_1-T_0)))*Q_1(y,0, t=t)
    return temp


y_bar_min = 0
y_bar_max = 1
dy_bar = 0.02
dt_bar = 0.00005
t = 1

num_y_bar_values = (y_bar_max - y_bar_min)/dy_bar + 1

y_bar_values = np.linspace(y_bar_min, y_bar_max, num_y_bar_values)
y_values = y_bar_values*120

T_bar = forward_euler(y_bar_min, y_bar_max, dy_bar, dt_bar, t, g, a, b, C_1, C_2)

T = T_bar*(T_1 - T_0) + T_0

t_1 = 0.05
t_2 = 0.15
t_3 = 0.5

t_1_index = int(t_1*t/dt_bar)
t_2_index = int(t_2*t/dt_bar)
t_3_index = int(t_3*t/dt_bar)

y_values = y_values - 120

plt.figure(figsize = (15,8))

plt.plot(y_values, T[0,:], alpha=0.9, linestyle="--", marker = "x", ms = 3)
plt.plot(y_values, T[t_1_index,:], alpha=0.9, linestyle="--", marker = "x", ms=5)
plt.plot(y_values, T[t_2_index,:], alpha=0.9, linestyle="--", marker = "x", ms=5)
plt.plot(y_values, T[t_3_index,:], alpha=0.9, linestyle="--", marker = "x", ms=5)
plt.plot(y_values, T[-1,:], alpha=0.9, linestyle="--", marker = "x", ms=3)

plt.title("Time evolution of temperature in the lithosphere", fontsize=20)

plt.ylabel(r"T $[^\circ C]$", fontsize=20)
plt.xlabel("y [km]", fontsize=20)

plt.legend(["t=0 Gy","t = %.3f Gy" %t_1,"t = %.3f Gy" %t_2,"t = %.3f Gy" %t_3,"t=%s Gy" %t, "Analytic"])

plt.show()
```

### It looks like the temperature increases with time until it reaches a steady state. The change is fast in the begining and then slows down. 

# We can now simulate the temperature with even more addtioional heat. Again, I assume that the initial condition is the linear function found as the steady state with no heat source. Half life of the radioactive elements are included.

### We see that the temperature increases in the beginning, but then decreaes after some time. This is a results of the half life of the radioactive elements.
